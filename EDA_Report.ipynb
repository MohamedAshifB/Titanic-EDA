```python
#Missing Values Handling
sns.heatmap(df.isnull(),cbar=False)
plt.title("Missing Data Heatmap")
plt.show()
```

```python
#filling the values and droping
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
df.drop(columns=['Cabin'],inplace=True)
```

```python
#Visualizations
#numerical
df.hist(figsize=(10,8))
plt.tight_layout()
plt.show()

sns.boxplot(x='Age',data=df)
```

```python
#Categorical
sns.countplot(x='Survived',data=df)
sns.countplot(x='Pclass',data=df)
```

```python
#Numerical vs Categorical
sns.boxplot(x='Survived',y='Age',data=df)
sns.violinplot(x='Survived',y='Fare', data=df)
```

```python
#Categorical vs Categorical
pd.crosstab(df['Survived'],df['Sex']).plot(kind='bar',stacked=True)
```

```python
#Correlation & Pairplots
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
sns.pairplot(df[['Survived','Age','Fare','Pclass']])
```

#insights & Inference 

1) Survival was more likely for females
2) 1st class passengers had better survival rates
3) children and young adults had higher survival probabilities
4) Fare and class are positively correlated
